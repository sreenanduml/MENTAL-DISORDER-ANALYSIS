```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print("\n Missing Values:")
print(m.isnull().sum())
```

```python
for col in m.select_dtypes(include='object').columns:
    m[col] = m[col].str.strip().str.upper()
```

```python
categorical_cols = m.select_dtypes(include='object').columns
for col in categorical_cols:
    print(f"\n Value counts for '{col}':")
    print(m[col].value_counts())
```

```python
df=m.copy()
df
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
m['Patient Number_e'] = label_encoder.fit_transform(m['Patient Number'])
m['Sadness_e'] = label_encoder.fit_transform(m['Sadness'])
m['Euphoric_e'] = label_encoder.fit_transform(m['Euphoric'])
m['Exhausted_e'] = label_encoder.fit_transform(m['Exhausted'])
m['Sleep dissorder_e'] = label_encoder.fit_transform(m['Sleep dissorder'])
m['Mood Swing_e'] = label_encoder.fit_transform(m['Mood Swing'])
m['Suicidal thoughts_e'] = label_encoder.fit_transform(m['Suicidal thoughts'])
m['Anorxia_e'] = label_encoder.fit_transform(m['Anorxia'])
m['Authority Respect_e'] = label_encoder.fit_transform(m['Authority Respect'])
m['Try-Explanation_e'] = label_encoder.fit_transform(m['Try-Explanation'])
m['Aggressive Response_e'] = label_encoder.fit_transform(m['Aggressive Response'])
m['Ignore & Move-On_e'] = label_encoder.fit_transform(m['Ignore & Move-On'])
m['Nervous Break-down_e'] = label_encoder.fit_transform(m['Nervous Break-down'])
m['Admit Mistakes_e'] = label_encoder.fit_transform(m['Admit Mistakes'])
m['Overthinking_e'] = label_encoder.fit_transform(m['Overthinking'])
m['Sexual Activity_e'] = label_encoder.fit_transform(m['Sexual Activity'])
m['Concentration_e'] = label_encoder.fit_transform(m['Concentration'])
m['Optimisim_e'] = label_encoder.fit_transform(m['Optimisim'])
m['Expert Diagnose_e'] = label_encoder.fit_transform(m['Expert Diagnose'])
m
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x='Expert Diagnose', data=m,hue='Expert Diagnose')
plt.title("Diagnosis Distribution")
plt.xlabel("Diagnosis Code")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
features = m.drop(columns=['Patient Number', 'Expert Diagnose'])

for col in features.columns:
    plt.figure(figsize=(5,3))
    sns.histplot(m[col], kde=False, bins=5)
    plt.title(f"Distribution: {col}")
    plt.tight_layout()
    plt.show()
```

```python
# Count plots for selected categorical features
for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=col, data=df)
    plt.title(f'Count Plot - {col}')
    plt.xticks(rotation=45)
    plt.show()
```
